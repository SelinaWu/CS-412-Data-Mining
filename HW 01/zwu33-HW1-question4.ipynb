```python
import numpy as np
import scipy
from scipy import spatial
# Line 3: 1: Manhattan distance
def M(target, ls):
    result = []
    for elem in ls:
        result.append(spatial.distance.cityblock(target,elem))
    return result
```

```python
# Line 3: 2: Euclidean distance
def E(target, ls):
    result = []
    for elem in ls:
        result.append(spatial.distance.minkowski(target,elem,2))
    return result
```

```python
# Line 3: 3: Supremum distance
def S(target, ls):
    result = []
    for elem in ls:
        result.append(spatial.distance.minkowski(target,elem,np.inf))
    return result
```

```python
# Line 3: 4: Cosine similarity
# from sklearn.metrics.pairwise import cosine_similarity
def C(target, ls):
    result = []
    for elem in ls:
        result.append(spatial.distance.cosine(target,elem))
    return result
```

```python
def distance(target,ls,idx):
    if(idx == 1):
        return M(target,ls)
    elif(idx == 2):
        return E(target,ls)
    elif(idx == 3):
        return S(target,ls)
    elif(idx == 4):
        return C(target,ls)

from sklearn.decomposition import PCA
from matplotlib import pyplot as plt

def pca(std):  
    sklearn_pca = PCA(n_components=X)
    sklearn_pca.fit(data_std)
    result = sklearn_pca.transform(data_std)
    var_exp = sklearn_pca.explained_variance_ratio_
    print("var_sum:", sum(var_exp))
    return result
```

```python
import copy

fp = 'Q4-analysis-input.in'
dataSet = []

raw_data = open(fp).read()
#raw_data = raw_data.split('\n')

raw_data = [n.split() for n in raw_data.split('\n')] 

for i in range(len(raw_data)):
    for j in range(len(raw_data[i])):
        raw_data[i][j] = int(raw_data[i][j])

D = raw_data[0][0]
N = raw_data[1][0]
distance_matrix = raw_data[2][0]
X = raw_data[3][0]

print("D: ",D,";\nN: ",N,";\ndistance_matrix:", distance_matrix,";\nX:",X)

for i in range(N + 1):
    if len(raw_data[i + 4])!= 0:
        dataSet.append(raw_data[i + 4])   

data_std = copy.deepcopy(dataSet)
```

```python
if(X != -1):
    #data_std = standardize(dataSet)
    data_std = pca(data_std)  

P = data_std[0] 
result = distance(P,data_std[1:],distance_matrix)
```

```python
# Make a list of (distance, index) tuples
result_pairs = [(result[i], i + 1) for i in range(len(result))]

# sort by distance
import heapq
heapq.heapify(result_pairs)

last_five = []
for i in range(5):
    last_five.append(heapq.heappop(result_pairs))
    
# write result to 'output.txt'
file = open('output','w') 

for i in range(5):
    file.write(str(last_five[i][1]))
    file.write('; ')
cum = sum(last_five[:5][0])
if(X != -1):
    file.write("\n")
    file.write(str("%.4f" % cum))
file.close()
```
