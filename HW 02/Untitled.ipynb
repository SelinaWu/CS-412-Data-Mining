```python
import sys
from itertools import combinations

partition = sys.stdin.readline()
raw_data = sys.stdin.read()

raw_data = [n.split() for n in raw_data.split('\n')] 
size = int(len(raw_data[0])/int(partition))
data = [] # fragments
dict_all = [] # value and inverted indexs, e.g. [{'a1': [1, 2, 3]}, {'b1': [3, 4, 5]}]


# split raw_data into fragment
c = 0
while c < int(len(raw_data[0])):
    temp = []
    for r in raw_data:
        temp.append(r[c: c+size]) 
    c = c + size    
    data.append(temp)
    
seq = [] # sequence at input
# check data freq.
for frag in data:
    dict_temp = {}
    for i in frag[0]: 
        seq.append(i[0])
    for ind in range(len(frag)):
        for k in frag[ind]:
            if dict_temp.get(k) == None: 
                dict_temp[k] = [ind]
            else:
                dict_temp[k].append(ind)
    dict_all.append(dict_temp)

# test = ['a1', 'a2', 'b1', 'b2', 'c1']
# [['a1', 'a2'], ['b1', 'b2'], ['c1']]
def set_by_char(ls):
    start = ls[0]
    test_after = []
    temp = []
    for i in ls:
        if i[0] == start[0]:
            temp.append(i)
        else:
            test_after.append(temp)
            temp = [i]
            start = i
    test_after.append(temp)
    return test_after


def comb(ls):
    comb_r = []
    num = 2
    while num <= len(ls):
        comb = combinations(ls, num)
        num += 1
        for i in comb:
            temp = []
            for j in i:
                temp.append(j)
            comb_r.append(temp)
    return comb_r

def second_comb(comb_r):
    comb_after = []
    for i in comb_r:
        temp = []
        for j in i:
            temp += j
        t = comb(temp)
        for k in t:
            if len(k) == len(i):
                comb_after.append(k)
    return comb_after

sequency = []
for i in seq:
    for frag in dict_all:
        for temp in sorted(frag.keys()):
            if i == temp[0]: 
                sequency.append(temp)
#print("sequency: ", sequency)

def intersect(a, b):
    return list(set(a) & set(b))

result = []
for dict_frag in dict_all:
    #print("dictionary: ",dict_frag)
    temp = sequency
    test = []
    for char in temp:
        if dict_frag.get(char) != None:
            output = char + " " + ':' + " " + str(len(dict_frag[char]))
            test.append(char)
            result.append(output)
    test = set_by_char(test)
    #print("test:",test)
    comb_r = comb(test)
    #print("combination:")
    comb_r = second_comb(comb_r)
    for elem in comb_r:
        t = dict_frag[elem[0]]
        for i in range(1,len(elem)):
            t = intersect(t,dict_frag[elem[i]])
        if len(t) > 0:
            s = ""
            for i in elem:
                s += i +" "
            output = s + ':' + " " + str(len(t))
            result.append(output)
    result.append("")
        
#print("result: ",result)
for r in result:
    sys.stdout.write(r)
    sys.stdout.write("\n")

```
